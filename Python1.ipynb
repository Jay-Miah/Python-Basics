```python
password = 123
user_entry = int(input("Enter your password:"))
print(user_entry == password)

while user_entry != password:
  print("Password is incorrect. Please try again.")

  user_entry = int(input("Enter your password:"))

  if user_entry == password:
    print("Password is correct.")
    break
```

```python
password = "abc123"
user_entry = input("Enter your password:")
print(user_entry == password)


if user_entry == password:
    print("Password is correct. Welcome to your account.")

else:
    attemps = 3
    while user_entry != password and attemps >0:
      print("Password is incorrect. Please try again.")

      user_entry = input("Enter your password:")
      if user_entry == password:
        print("Password is correct.")
        break
      attemps -=1
```
